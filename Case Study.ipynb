## Importing the dataset

```python
# reading the file 
import pandas as pd
pd.set_option('display.float_format', lambda x: '%.0f' % x)

case_data1 = pd.read_excel("Input_data1.xlsx")
case_data1 = case_data1.dropna(subset= ['ClearingDocNumber'],axis = 0)
case_data1
```

## Invoice Dataframe

```python
invoice_df = case_data1[case_data1["doc_type"] == "RV"]
invoice_df
```

## Payment Dataframe

```python
req_doc = ["ZP", "SB"]
payment_df = case_data1[case_data1.doc_type.isin(req_doc)]
payment_df
```

## Filtering DocNum == ClearingNum in Payment Dataframe

```python
e_payment_df = payment_df[payment_df.Document_Num == payment_df.ClearingDocNumber]
e_payment_df
```

## Merging two dataset on ClearingDocNumber when DocNum == ClearingNum in Payment Dataframe

```python
e_merged_df = pd.merge(invoice_df,e_payment_df,left_on="ClearingDocNumber",right_on="Document_Num",how = 'inner')
e_merged_df
```

## Filtering DocNum != ClearingNum in Payment Dataframe

```python
u_payment_df = payment_df[payment_df.Document_Num != payment_df.ClearingDocNumber]
u_payment_df
```

## Merging two dataset on ClearingDocNumber when DocNum != ClearingNum in Payment Dataframe

```python
u_merged_df = pd.merge(invoice_df,u_payment_df,left_on="ClearingDocNumber",right_on = "Document_Num",how = 'inner')
u_merged_df
```

# 1. Identify the invoices that are on-time, not on-time, open w.r.t i) payment date ii) clearing date


## i) Payment Time 

## Calculating on-time invoice

```python
on_time_pay = e_merged_df[e_merged_df.due_date_x >= e_merged_df.posting_date_y]
len(on_time_pay)
```

## Calculating not on-time invoice

```python
off_time_pay = e_merged_df[e_merged_df.due_date_x < e_merged_df.posting_date_y]
len(off_time_pay)
```

## Calculating Open time invoice

```python
4111 - 3981
```

## ii) Clearing Time

## Calculating on-time invoice

```python
on_time_cl = e_merged_df[e_merged_df.due_date_x >= e_merged_df.clearingdate_x]

u_on_time_cl = u_merged_df[u_merged_df.due_date_x >= u_merged_df.clearingdate_x]

total_on_time_cl = len(on_time_cl) + len(u_on_time_cl)
total_on_time_cl
```

## Calculating off-time invoice

```python
off_time_cl = len(e_merged_df) - len(on_time_cl)

u_off_time_cl = len(u_merged_df) - len(u_on_time_cl)

total_off_time_cl = off_time_cl + u_off_time_cl
total_off_time_cl
```

## Calculating Open time invoice

```python
(4111 - 3981) + 24
```

## Function to Plot Bar graph


```python
import matplotlib.pyplot as plt

def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])
        
def bar_plot(x_data,y_data,x_label,y_label,title):
    plt.bar(x_data,y_data)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    addlabels(x_data,y_data)
    
    plt.show()
```
